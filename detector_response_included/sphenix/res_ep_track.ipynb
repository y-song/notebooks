```python
import time
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
%run functions.ipynb
```

```python
x = [0.5, 1.5, 3.5, 7.5, 15]
xerr = [0.5, 0.5, 1.5, 2.5, 5]
y = [7.31706e-3, 6.4916e-3, 7.93033e-3, 1.31878e-2, 1.78008e-2]
yerr = [1.06456e-4, 8.28384e-5, 1.11556e-4, 3.09644e-4, 5.52370e-4]
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(x,y, 'ro')
plt.errorbar(x, y, xerr=xerr, yerr=yerr, color='r', fmt='none')
plt.title('Tracks in ep', fontsize=15)
plt.xlabel('$p^{tru}$ (GeV)', fontsize=15)
plt.ylabel(r'$\sigma(\frac{dp}{p})$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(prop={'size': 15}, frameon=False, loc='upper center')
plt.xlim(0,20)
plt.savefig('images/res_ep_track.png', bbox_inches='tight')
```
