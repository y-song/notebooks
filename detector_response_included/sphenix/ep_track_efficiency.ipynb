```python
import time
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
from statsmodels.stats.proportion import proportion_confint
```

```python
table = np.loadtxt('uploads/eff_p_2.txt')
n = table[:,0]
nRec = table[:,1]
```

```python
y = []
yerr = []
for i in range(0, len(n)):
    if n[i] != 0:
        y.append(nRec[i] / n[i])
        yerr.append(abs(proportion_confint(nRec[i], n[i], 1 - 0.68, 'jeffreys') - nRec[i] / n[i]))
    else:
        y.append(0)
        yerr.append(np.asarray([0,0]))
```

```python
x = np.arange(-1.05, 1.15, 0.1)
```

```python
fig = plt.figure(figsize=(10,8))
plt.errorbar(x, y, xerr=0.05, yerr=np.transpose(np.asarray(yerr)), label='')
plt.xlabel('$\eta^{tru}$', fontsize=15)
plt.ylabel('Efficiency', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('ep charged track reconstruction efficiency in sPHENIX, $p^{tru}$ > 2 GeV', fontsize=15)
plt.ylim(0, 1)
plt.text(-1.1, 0.9, ' computed using information from "ntp_gtrack", \n cuts: $|\Delta p / p^{tru}| < 0.03, |\Delta \eta| < 0.03, |\Delta \phi| < 0.03$', fontsize=15)
plt.savefig('images/track_eff_vs_eta_p_2.png', bbox_inches='tight')
```
