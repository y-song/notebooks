```python
import time
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
%run functions.ipynb
```

Detector setup is the same as the screenshot in "Group Meeting Presentation 011020" (all Si). See github commits on 011520 for simulation.C (default number of events = 100000)

```python
def getData(p):
    y = np.asarray(np.loadtxt('uploads/res_vs_eta_y/2_gaus/0305/'+str(p)+'.txt'))
    yerr = np.asarray(np.loadtxt('uploads/res_vs_eta_yerr/2_gaus/0305/'+str(p)+'.txt'))
    return y, yerr
```

```python
binEdge = np.arange(-5.0, 5.2, 0.2)
x = []
for i in range(len(binEdge)-1):
    x.append((binEdge[i]+binEdge[i+1])/2)
```

```python
fig = plt.figure(figsize=(12,6))
#plt.errorbar(x, getData(20)[0], xerr=0.1, yerr=getData(20)[1], fmt='none', label='$p^{tru} = 20 \, GeV$')
plt.errorbar(x, getData(20)[0], xerr=0.1, yerr=getData(20)[1], fmt='none', label='$p^{tru} = 20 \, GeV$')
plt.errorbar(x, getData(10)[0], xerr=0.1, yerr=getData(10)[1], fmt='none', label='$p^{tru} = 10 \, GeV$')
plt.errorbar(x, getData(5)[0], xerr=0.1, yerr=getData(5)[1], fmt='none', label='$p^{tru} = 5 \, GeV$')
#plt.errorbar(x, getData(1)[0], xerr=0.1, yerr=getData(1)[1], fmt='none', label='$p^{tru} = 1 \, GeV$')
plt.title('Single $\pi^-$ tracks, 100,000 events for each $p^{tru}$', fontsize=15)
plt.xlim(left=-3.2, right=3.2)
plt.ylim(bottom=0, top=0.05)
plt.xlabel('$\eta^{tru}$', fontsize=15)
plt.ylabel(r'$\sigma(\frac{dp}{p})$', fontsize=15)
plt.xticks(np.arange(-3.0, 3.2, 1.0), fontsize=14)
plt.yticks(fontsize=15)
plt.legend(prop={'size': 15}, frameon=False, loc='upper center')
plt.savefig('images/res_2gaus_0305.png', bbox_inches='tight')
```
