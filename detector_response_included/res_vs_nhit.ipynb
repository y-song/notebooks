```python
import time
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
%run functions.ipynb
```

```python
def getData(p):
    y = np.asarray(np.loadtxt('uploads/res_vs_nhit/'+str(p)+'.txt'))
    yerr = np.asarray(np.loadtxt('uploads/res_vs_nhit_err/'+str(p)+'.txt'))
    return y, yerr
```

```python
x = np.arange(3,9)
print x
```

```python
fig = plt.figure(figsize=(12,6))
plt.errorbar(x, getData(20)[0], yerr=getData(20)[1], label='$p^{tru} = 20 \, GeV$')
plt.errorbar(x, getData(5)[0], yerr=getData(5)[1], label='$p^{tru} = 5 \, GeV$')
plt.errorbar(x[:-1], getData(1)[0], yerr=getData(1)[1], label='$p^{tru} = 1 \, GeV$')
plt.title('Single $\pi^-$ tracks, 100,000 events for each $p^{tru}$ and $1.0 < \eta < 1.2$', fontsize=15)
plt.xlabel('number of hits', fontsize=15)
plt.ylabel(r'$\sigma(\frac{dp}{p})$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(prop={'size': 15}, frameon=False, loc='upper center')
plt.savefig('images/res_vs_nhit1.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(12,6))
plt.errorbar(x, getData(20)[0], yerr=getData(20)[1], label='$p^{tru} = 20 \, GeV$')
plt.errorbar(x, getData(5)[0], yerr=getData(5)[1], label='$p^{tru} = 5 \, GeV$')
plt.errorbar(x[:-1], getData(1)[0], yerr=getData(1)[1], label='$p^{tru} = 1 \, GeV$')
plt.title('Single $\pi^-$ tracks, 100,000 events for each $p^{tru}$ and $1.0 < \eta < 1.2$', fontsize=15)
plt.xlabel('number of hits', fontsize=15)
plt.ylabel(r'$\sigma(\frac{dp}{p})$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.yscale('log')
plt.legend(prop={'size': 15}, frameon=False, loc='upper center')
plt.savefig('images/res_vs_nhit2.png', bbox_inches='tight')
```
