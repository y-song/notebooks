```python
import time
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
from statsmodels.stats.proportion import proportion_confint
%run functions.ipynb
```

Detector setup is the same as the screenshot in "Group Meeting Presentation 011020" (all Si). See github commits on 011520 for simulation.C (default number of events = 100000) and analysis_fit.C code. eta_tru_arr1-20.txt and eta_rec_arr1-20.txt are cout results from analysis_fit.C. See Yue Shi's email for details on proportion_confint function.

```python
def getData(p):
    nEvt = np.asarray(np.loadtxt('uploads/eta_tru_arr/'+str(p)+'.txt'))
    nRec = np.asarray(np.loadtxt('uploads/eta_rec_arr/'+str(p)+'.txt'))
    return nEvt, nRec
```

```python
def calc(p):
    nEvt, nRec = getData(p)
    y = nRec / nEvt
    yerr = abs(proportion_confint(nRec, nEvt, 1 - 0.68, 'jeffreys') - y)
    return y, yerr
```

```python
x = np.arange(-5.0, 5.0, 0.1)
```

```python
fig = plt.figure(figsize=(10,8))
plt.errorbar(x, calc('20-')[0], xerr=0.05, yerr=calc('20-')[1], label='$p^{tru} = 20 \, GeV$')
plt.errorbar(x, calc('10-')[0], xerr=0.05, yerr=calc('10-')[1], label='$p^{tru} = 10 \, GeV$')
plt.errorbar(x, calc('5-')[0], xerr=0.05, yerr=calc('5-')[1], label='$p^{tru} = 5 \, GeV$')
plt.errorbar(x, calc('1-')[0], xerr=0.05, yerr=calc('1-')[1], label='$p^{tru} = 1 \, GeV$')
plt.title('Single $\pi^-$ tracks', fontsize=15)
plt.xlim(left=-5.0, right=5.0)
plt.xlabel('$\eta^{tru}$', fontsize=15)
plt.ylabel('Efficiency', fontsize=15)
plt.xticks(np.arange(-5.0, 6.0, 1.0), fontsize=15)
plt.yticks(fontsize=15)
plt.legend(prop={'size': 15}, frameon=False, loc='best')
plt.savefig('images/eff_stack_new.png', bbox_inches='tight')
```
